### Cleaning Data

Membersihkan Missing Value dari `aothizhongxin_dframe`

```python
aothizhongxin_dframe["PM2.5"].fillna(value=aothizhongxin_dframe["PM2.5"].mean(), inplace=True)
aothizhongxin_dframe.PM10.fillna(value=aothizhongxin_dframe.PM10.mean(), inplace=True)
aothizhongxin_dframe.SO2.fillna(value=aothizhongxin_dframe.SO2.mean(), inplace=True)
aothizhongxin_dframe.NO2.fillna(value=aothizhongxin_dframe.NO2.mean(), inplace=True)
aothizhongxin_dframe.CO.fillna(value=aothizhongxin_dframe.CO.mean(), inplace=True)
aothizhongxin_dframe.O3.fillna(value=aothizhongxin_dframe.O3.mean(), inplace=True)
aothizhongxin_dframe.TEMP.fillna(value=aothizhongxin_dframe.TEMP.mean(), inplace=True)
aothizhongxin_dframe.PRES.fillna(value=aothizhongxin_dframe.PRES.mean(), inplace=True)
aothizhongxin_dframe.DEWP.fillna(value=aothizhongxin_dframe.DEWP.mean(), inplace=True)
aothizhongxin_dframe.RAIN.fillna(value=aothizhongxin_dframe.RAIN.mean(), inplace=True)
aothizhongxin_dframe.wd.fillna(value="NE", inplace=True)
aothizhongxin_dframe.WSPM.fillna(value=aothizhongxin_dframe.WSPM.mean(), inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun mana yang memiliki suhu paling panas?

```python
temp_by_station = all_df.groupby('station')['TEMP'].max().reset_index()
temp_by_station.sort_values(by='TEMP',ascending=False, inplace=True)
plt.figure(figsize=(12, 8))
sns.barplot(
    y='station', 
    x='TEMP', 
    data=temp_by_station, 
    palette='Blues_d', 
    ci=None  
)
plt.title("Stasiun Dengan Temperature Terpanas", fontsize=16)
plt.xlabel("Temperature", fontsize=14)
plt.ylabel("Staisun", fontsize=14)
plt.show()
```

### Pertanyaan 2: Stasiun mana yang memiliki jumlah polusi paling banyak pada 2017

```python
temp_by_station2 = all_df.groupby('station')['RAIN'].max().reset_index()
temp_by_station2.sort_values(by='RAIN',ascending=False, inplace=True)
plt.figure(figsize=(12, 8))
sns.barplot(
    y='station', 
    x='RAIN', 
    data=temp_by_station2, 
    palette='Blues_d', 
    ci=None  
)
plt.title("Stasiun Dengan Curah Hujan Tertinggi", fontsize=16)
plt.xlabel("Curah Hujan", fontsize=14)
plt.ylabel("Stasiun", fontsize=14)
plt.show()
```
